# 01 — Coleta de Dados

Fontes utilizadas:
- **BCB/SGS** — Câmbio USD/BRL, Selic e IPCA via API REST pública
- **CEPEA/ESALQ** — Preços diários de soja e milho (planilha Excel)

> **Nota CEPEA:** O site não oferece API pública. O fluxo recomendado é:
> 1. Baixar o `.xlsx` manualmente em https://www.cepea.esalq.usp.br/br/indicador/
> 2. Salvar em `data/raw/` com os nomes `cepea_soja.xlsx` e `cepea_milho.xlsx`
> 3. Rodar a célula de leitura abaixo
>
> A função `fetch_cepea_file()` tenta automatizar via `requests`, mas pode falhar
> se o site exigir cookies de sessão — o download manual é sempre o fallback confiável.

```python
import time
import logging
from pathlib import Path
from datetime import date, timedelta

import requests
import pandas as pd

logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s [%(levelname)s] %(message)s',
    datefmt='%H:%M:%S',
)
log = logging.getLogger(__name__)

RAW_DIR = Path('../data/raw')
RAW_DIR.mkdir(parents=True, exist_ok=True)

print('Diretório de dados brutos:', RAW_DIR.resolve())
```

---
## 1. Banco Central — Séries macro (API SGS)

```python
BCB_BASE_URL = 'https://api.bcb.gov.br/dados/serie/bcdata.sgs.{serie}/dados'

# Séries relevantes para o projeto
# Série 1      → Câmbio USD/BRL (compra, diário)
# Série 21619  → Selic Over (% a.a.) — custo de capital e carregamento de estoque
# Série 433    → IPCA (% ao mês) — deflação de série histórica de preços
BCB_SERIES = {
    'usd_brl': 1,
    'selic':   21619,
    'ipca':    433,
}


def fetch_bcb_series(
    serie_code: int,
    start_date: str,
    end_date: str,
    max_retries: int = 3,
    backoff: float = 2.0,
) -> pd.DataFrame:
    """
    Baixa uma série temporal do SGS/BCB.

    Args:
        serie_code : Código numérico da série no SGS.
        start_date : Data inicial no formato 'DD/MM/YYYY'.
        end_date   : Data final   no formato 'DD/MM/YYYY'.
        max_retries: Tentativas em caso de falha HTTP.
        backoff    : Fator de espera entre tentativas (segundos).

    Returns:
        DataFrame com colunas ['data', 'valor'].
    """
    url = BCB_BASE_URL.format(serie=serie_code)
    params = {
        'formato':      'json',
        'dataInicial':  start_date,
        'dataFinal':    end_date,
    }

    for attempt in range(1, max_retries + 1):
        try:
            log.info('BCB série %s — tentativa %d/%d', serie_code, attempt, max_retries)
            resp = requests.get(url, params=params, timeout=30)
            resp.raise_for_status()
            data = resp.json()

            if not data:
                log.warning('BCB retornou lista vazia para série %s.', serie_code)
                return pd.DataFrame(columns=['data', 'valor'])

            df = pd.DataFrame(data)
            df['data']  = pd.to_datetime(df['data'], dayfirst=True)
            df['valor'] = pd.to_numeric(df['valor'], errors='coerce')
            df = df.sort_values('data').reset_index(drop=True)
            log.info('BCB série %s — %d registros baixados.', serie_code, len(df))
            return df

        except requests.exceptions.HTTPError as exc:
            log.error('HTTP %s para série %s: %s', exc.response.status_code, serie_code, exc)
        except requests.exceptions.RequestException as exc:
            log.error('Erro de rede (tentativa %d): %s', attempt, exc)

        if attempt < max_retries:
            wait = backoff ** attempt
            log.info('Aguardando %.1fs antes da próxima tentativa...', wait)
            time.sleep(wait)

    log.error('Falha ao baixar série %s após %d tentativas.', serie_code, max_retries)
    return pd.DataFrame(columns=['data', 'valor'])
```

```python
# 5 anos de histórico — suficiente para capturar ciclos de safra completos
END_DATE   = date.today()
START_DATE = END_DATE - timedelta(days=5 * 365)

start_str = START_DATE.strftime('%d/%m/%Y')
end_str   = END_DATE.strftime('%d/%m/%Y')

print(f'Período: {start_str} → {end_str}')

bcb_dfs = {}

for nome, codigo in BCB_SERIES.items():
    df = fetch_bcb_series(codigo, start_str, end_str)
    if not df.empty:
        df = df.rename(columns={'valor': nome})
        bcb_dfs[nome] = df
        out_path = RAW_DIR / f'bcb_{nome}.parquet'
        df.to_parquet(out_path, index=False)
        log.info('Salvo: %s', out_path)

# Consolida em uma única tabela (outer join por data)
if bcb_dfs:
    df_bcb = list(bcb_dfs.values())[0][['data']].copy()
    for nome, df in bcb_dfs.items():
        df_bcb = df_bcb.merge(df[['data', nome]], on='data', how='outer')
    df_bcb = df_bcb.sort_values('data').reset_index(drop=True)
    df_bcb.to_parquet(RAW_DIR / 'bcb_macro.parquet', index=False)
    print(f'\nTabela macro consolidada: {df_bcb.shape}')
    df_bcb.tail()
```

```python
print(df_bcb.dtypes)
print('\nNulos por coluna:')
print(df_bcb.isnull().sum())
print('\nEstatísticas básicas:')
df_bcb.describe()
```

## 2. CEPEA/ESALQ — Preços de Soja e Milho

### Opção A — Download manual (recomendado, sempre funciona)

1. Acesse https://www.cepea.esalq.usp.br/br/indicador/soja.aspx  
   Clique em **"Planilha de dados"** e baixe o `.xlsx`
2. Repita para milho: https://www.cepea.esalq.usp.br/br/indicador/milho.aspx
3. Salve como `data/raw/cepea_soja.xlsx` e `data/raw/cepea_milho.xlsx`
4. Execute as células de leitura abaixo

### Opção B — Download automatizado (tenta primeiro, pode falhar)
O CEPEA não oferece API pública; o scraping abaixo pode exigir cookies de sessão.

```python
from html.parser import HTMLParser

CEPEA_PAGES = {
    'soja':  'https://www.cepea.esalq.usp.br/br/indicador/soja.aspx',
    'milho': 'https://www.cepea.esalq.usp.br/br/indicador/milho.aspx',
}

HEADERS = {
    'User-Agent': (
        'Mozilla/5.0 (Windows NT 10.0; Win64; x64) '
        'AppleWebKit/537.36 (KHTML, like Gecko) '
        'Chrome/120.0.0.0 Safari/537.36'
    ),
    'Referer': 'https://www.cepea.esalq.usp.br/',
}


class XlsxLinkParser(HTMLParser):
    """Extrai links para arquivos .xlsx/.xls de uma página HTML."""

    def __init__(self):
        super().__init__()
        self.xlsx_links = []

    def handle_starttag(self, tag, attrs):
        if tag == 'a':
            href = dict(attrs).get('href', '')
            if '.xlsx' in href or '.xls' in href:
                self.xlsx_links.append(href)


def fetch_cepea_file(commodity: str, save_path: Path) -> bool:
    """
    Tenta baixar o Excel do CEPEA via scraping.
    Retorna True se bem-sucedido, False caso contrário.
    """
    session = requests.Session()
    session.headers.update(HEADERS)

    try:
        log.info('CEPEA: acessando página de %s...', commodity)
        resp = session.get(CEPEA_PAGES[commodity], timeout=30)
        resp.raise_for_status()

        parser = XlsxLinkParser()
        parser.feed(resp.text)

        if not parser.xlsx_links:
            log.warning(
                'CEPEA: nenhum link .xlsx encontrado para %s.\n'
                '→ Faça o download manual em %s',
                commodity, CEPEA_PAGES[commodity],
            )
            return False

        xlsx_url = parser.xlsx_links[0]
        if not xlsx_url.startswith('http'):
            xlsx_url = 'https://www.cepea.esalq.usp.br' + xlsx_url

        log.info('CEPEA: baixando %s em %s', commodity, xlsx_url)
        file_resp = session.get(xlsx_url, timeout=60)
        file_resp.raise_for_status()

        with open(save_path, 'wb') as f:
            f.write(file_resp.content)

        log.info('CEPEA: salvo em %s (%.1f KB)', save_path, len(file_resp.content) / 1024)
        return True

    except requests.exceptions.RequestException as exc:
        log.error('CEPEA: falha no download de %s: %s', commodity, exc)
        return False


for commodity in ['soja', 'milho']:
    dest = RAW_DIR / f'cepea_{commodity}.xlsx'
    if dest.exists():
        log.info('Arquivo já existe: %s — pulando download.', dest.name)
        continue
    success = fetch_cepea_file(commodity, dest)
    if not success:
        print(f'\n⚠️  Download automático falhou para {commodity}.')
        print(f'   → Baixe manualmente e salve em: {dest.resolve()}')
```

```python
def parse_cepea_xlsx(path: Path, commodity: str) -> pd.DataFrame:
    """
    Lê e padroniza o Excel do CEPEA.

    O layout do arquivo tem metadados nas primeiras linhas e o cabeçalho
    real em posição variável — a função detecta automaticamente.

    Returns:
        DataFrame com colunas: data, preco_rs, preco_usd (opcional), commodity.
    """
    if not path.exists():
        log.error('Arquivo não encontrado: %s', path)
        return pd.DataFrame()

    # Detecta linha do cabeçalho real (contém a palavra 'data')
    raw = pd.read_excel(path, header=None, nrows=10)
    header_row = 0
    for i, row in raw.iterrows():
        if row.astype(str).str.lower().str.contains('data').any():
            header_row = i
            break

    df = pd.read_excel(path, header=header_row)
    df.columns = df.columns.str.strip().str.lower()
    log.info('Colunas encontradas (%s): %s', commodity, df.columns.tolist())

    # Mapeamento flexível de colunas
    col_map = {}
    for col in df.columns:
        if 'data' in col:
            col_map[col] = 'data'
        elif any(k in col for k in ['r$', 'reais', 'à vista']):
            col_map[col] = 'preco_rs'
        elif any(k in col for k in ['us$', 'dólar', 'usd']):
            col_map[col] = 'preco_usd'

    df = df.rename(columns=col_map)
    cols_keep = [c for c in ['data', 'preco_rs', 'preco_usd'] if c in df.columns]
    df = df[cols_keep].copy()

    # Parsing de data
    df['data'] = pd.to_datetime(df['data'], dayfirst=True, errors='coerce')
    df = df.dropna(subset=['data'])

    # Parsing de preços — trata vírgula como separador decimal (padrão BR)
    for col in ['preco_rs', 'preco_usd']:
        if col in df.columns:
            if df[col].dtype == object:
                df[col] = (
                    df[col]
                    .str.replace('.', '', regex=False)
                    .str.replace(',', '.', regex=False)
                )
            df[col] = pd.to_numeric(df[col], errors='coerce')

    df['commodity'] = commodity
    df = df.sort_values('data').reset_index(drop=True)

    log.info(
        'CEPEA %s: %d registros | %s → %s',
        commodity, len(df),
        df['data'].min().date(),
        df['data'].max().date(),
    )
    return df


cepea_dfs = {}

for commodity in ['soja', 'milho']:
    path = RAW_DIR / f'cepea_{commodity}.xlsx'
    df = parse_cepea_xlsx(path, commodity)
    if not df.empty:
        cepea_dfs[commodity] = df
        out_path = RAW_DIR / f'cepea_{commodity}.parquet'
        df.to_parquet(out_path, index=False)
        log.info('Salvo: %s', out_path)
        display(df.head(3))
```

---
## 3. Consolidação: CEPEA + câmbio

```python
# Junta preços CEPEA com câmbio BCB
# Útil para: (a) verificar cobertura temporal, (b) calcular preco_usd quando ausente

if cepea_dfs and 'usd_brl' in bcb_dfs:
    df_cambio = bcb_dfs['usd_brl'][['data', 'usd_brl']]

    for commodity, df_cepea in cepea_dfs.items():
        df_merged = df_cepea.merge(df_cambio, on='data', how='left')

        # Calcula preço em USD via câmbio quando coluna não vem na planilha
        if 'preco_usd' not in df_merged.columns or df_merged['preco_usd'].isna().all():
            df_merged['preco_usd_calc'] = df_merged['preco_rs'] / df_merged['usd_brl']

        out_path = RAW_DIR / f'cepea_{commodity}_cambio.parquet'
        df_merged.to_parquet(out_path, index=False)

        print(f'\n{commodity.upper()} + câmbio:')
        print(f'  Registros  : {len(df_merged)}')
        print(f'  Nulos preco_rs : {df_merged["preco_rs"].isna().sum()}')
        print(f'  Nulos usd_brl  : {df_merged["usd_brl"].isna().sum()}')
        display(df_merged.tail(3))

print('\n📦 Arquivos gerados em data/raw/:')
for f in sorted(RAW_DIR.glob('*.parquet')):
    size_kb = f.stat().st_size / 1024
    n = len(pd.read_parquet(f))
    print(f'  {f.name:<45} {n:>6} registros  ({size_kb:.1f} KB)')

print('\n✅ Coleta concluída. Próximo passo: 02_eda.ipynb')
```
